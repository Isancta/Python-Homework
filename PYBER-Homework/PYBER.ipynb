```python
#Import pandas for csv reading,Numpy for calculations and matplotlib for charting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Path to data
city_data_pd = "city_data.csv"
ride_data_pd = "ride_data.csv"
```

```python
# Read with pandas
city_info = pd.read_csv(city_data_pd)
ride_info = pd.read_csv(ride_data_pd)
city_info.head()
ride_info.head()
```

```python
pyber_data=pd.merge(ride_info,city_info,on=["city",], how="left")
pyber_data.head()
```

```python
#Review by Urban Cities
urban_cities = pyber_data[pyber_data["type"] == "Urban"]
urban_average_fare=urban_cities.groupby(["city"]).mean()["fare"]
urban_number_rider=urban_cities.groupby(["city"]).count()["ride_id"]
urban_number_drivers=urban_cities.groupby(["city"]).count()["driver_count"]
urban_cities.head()
```

```python
#Review by Suburban Cities
suburban_cities = pyber_data[pyber_data["type"] == "Suburban"]
suburban_number_fare=suburban_cities.groupby(["city"]).mean()["fare"]
suburban_number_rider=suburban_cities.groupby(["city"]).count()["ride_id"]
suburban_number_drivers=suburban_cities.groupby(["city"]).count()["driver_count"]
```

```python
#Review by Rural Cities
Rural_cities = pyber_data[pyber_data["type"] == "Rural"]
Rural_number_fare=Rural_cities.groupby(["city"]).mean()["fare"]
Rural_number_rider=Rural_cities.groupby(["city"]).count()["ride_id"]
Rural_number_drivers=Rural_cities.groupby(["city"]).count()["driver_count"]
```

```python
#Bubble chart by City type
plt.scatter(urban_number_rider,urban_average_fare, s=10*urban_number_drivers,label="Urban", facecolors="coral",edgecolor="black", linewidths=1, marker="o",
           alpha=0.75)
plt.scatter(suburban_number_rider,suburban_number_fare,s=10*suburban_number_drivers,label="Suburban",facecolors="skyblue",edgecolor="black", linewidths=1, marker="o",alpha=0.75)

plt.scatter(Rural_number_rider,Rural_number_fare,s=10*Rural_number_drivers,label="Rural",facecolors="Gold",edgecolor="black", linewidths=1, marker="o",alpha=0.75)

plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides(Per City)")
plt.ylabel("Average fare($)")

plt.text(42,42, "Notes:\n Circle size correlates with driver count per city")
plt.grid()
legnd=plt.legend(title="City Types",labels=["Urban","Suburban","Rural"],loc= "best",fontsize="medium",mode="expanded",numpoints=5,scatterpoints=1)
legnd.legendHandles[0]._sizes=[30]
legnd.legendHandles[1]._sizes=[30]
legnd.legendHandles[2]._sizes=[30]
plt.show()
```

```python
plt.savefig("../data/pyber bubble plot.png")
plt.show()
```

```python
#Pie chart parameters
Total_fare=pyber_data.groupby(["type"]).sum()["fare"]

Explode=(0.1,0.05,0.05)
colors=["coral","lightskyblue","gold"]
label=["Rural","Suburban","Urban"]
plt.pie(Total_fare, explode=Explode, labels=label, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title(" % of Total Fares by City Type")

plt.show()           
```

```python
Total_rides=pyber_data.groupby(["type"]).sum()["ride_id"]

Explode=(0.05,0.05,0.05)
colors=["coral","lightskyblue","gold"]
label=["Rural","Suburban","Urban"]
plt.pie(Total_rides, explode=Explode, labels=label, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type")

plt.show()   
```

```python
Total_drivers=pyber_data.groupby(["type"]).sum()["driver_count"]

Explode=(0.2,0.2,0.2)
colors=["coral","lightskyblue","gold"]
label=["Rural","Suburban","Urban"]
plt.pie(Total_drivers, explode=Explode, labels=label, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Drivers by City Type")

plt.show()   
```

```python
print("Comment:From data observation through various charts;Urban areas tend to have more drivers than rides which lower the riding fares,contrary to Rural areas where riding fares are significantly higher due to lack of drivers mostly;the bubble chart illustrate it clearly.The Suburbs have approximately a balance of drivers and rides, hense a relatively low riding fare.")
```
